## Notebook imports and packages

```python
import matplotlib.pyplot as plt 
import numpy as np 

%matplotlib inline
```

## Example 1

## $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make data 
x_1 = np.linspace(start=-3,stop=3,num=500)
```

```python
#Plot
plt.xlim([-3,3])
plt.ylim([0,8])

plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)

plt.plot(x_1,f(x_1))
plt.show()
```
